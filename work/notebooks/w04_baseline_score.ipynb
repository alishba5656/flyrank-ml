## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Simple baseline rule
df["score"] = (
    (df["ctr"] < 0.20).astype(int) +
    (df["avg_position"] > 20).astype(int)
)

df["reason_code"] = "CTR_FIX"
df["action"] = "Improve SEO"

df[["content_id","ctr","avg_position","score","reason_code","action"]].head(10)
```

### Baseline Rule

My baseline rule focuses on pages with low CTR and poor average ranking position.

Rule:
- CTR < 0.20
- Average Position > 20

Reason Code:
CTR_FIX

Action:
Improve SEO by optimizing titles, meta descriptions, and page content.

This rule is intended as a simple baseline that can later be compared with a machine learning model.

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import os

baseline = df.sort_values("score", ascending=False)

os.makedirs("work/outputs", exist_ok=True)

baseline.to_csv(
    "work/outputs/baseline_action_score.csv",
    index=False
)

baseline.head(20)
```

## 3. Top-20 review

*For each of the top 20: action, reason code, confidence note, and what would make it wrong.*

```python
top20 = baseline.head(20)

for i, row in top20.iterrows():
    print(f"""
Content: {row['content_id']}

Action:
Improve SEO

Reason:
Low CTR ({row['ctr']}) and Avg Position ({row['avg_position']}).

What would make it wrong?
The page may target a low-volume keyword, be seasonal, or have recently improved.
""")
```

## 4. Weak picks + leakage check

*Which picks look wrong and why? Confirm no product flags or future windows leaked in.*

```python
print("Weak Picks")

weak = baseline[baseline["score"] == 1].head(10)

weak[["content_id","ctr","avg_position","score"]]
```

### Weak Picks
